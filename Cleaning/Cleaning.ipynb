```python
# keep the columns we need for this project
data = dataold.filter(
    [
        "id",
        "listing_url",
        "name",
        "picture_url",
        "host_id",
        "host_name",
        "host",
        "verifications",
        "neighbourhood_group_cleansed",
        "latitude",
        "longitude",
        "room_type",
        "accommodates",
        "amenities",
        "price",
        "number_of_reviews",
        "review_scores_rating",
        "listing_count",
    ],
    axis=1,
)
```

```python
data.shape
```

```python
data.info()
```

```python
# make the price column from object to int
data["price"].replace(to_replace="^\$", value="", inplace=True, regex=True)
data["price"].replace(to_replace="\,", value="", inplace=True, regex=True)
data["price"] = pd.to_numeric(data["price"])
```

```python
data["price"]
```

```python
data.describe()
```

```python
# the amenities column is an object, i want to make a list from the amenities
# first step clean the punctuations [] and "
data["amenities"].replace(to_replace="^\[", value="", inplace=True, regex=True)
data["amenities"].replace(to_replace="\]", value="", inplace=True, regex=True)
data["amenities"].replace(to_replace='"', value="", inplace=True, regex=True)
```

```python
data["amenities"]
```

```python
# second step i make a copy of the dataset
amenities = data.copy()
# remove the nulls of amenities column
amenities = amenities[amenities["amenities"].notnull()]
```

```python
# now i convert the string to list
amenities["amenities"] = amenities["amenities"].apply(lambda a: a.split(","))
```

```python
# i want to count the amenities for all the airbnbs and i make a new column amenity
s = (
    amenities.apply(lambda x: pd.Series(x["amenities"]), axis=1)
    .stack()
    .reset_index(level=1, drop=True)
)
s.name = "amenity"

amenities = amenities.drop("amenities", axis=1).join(s)
```
